# Optimizing: Dealing with outliers and applying log

```python
# Define a custom transformer to handle outliers
def handle_outliers(X):
    X = X.copy()  # Avoid changes to original data
    for column in X.columns:
        Q1 = X[column].quantile(0.25)
        Q3 = X[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        X[column] = np.where(X[column] < lower_bound, lower_bound, X[column])
        X[column] = np.where(X[column] > upper_bound, upper_bound, X[column])
    return X

# Transformer to apply logarithmic transformation
def log_transform(X):
    return np.log1p(X)

outlier_transformer = FunctionTransformer(handle_outliers)
log_transformer = FunctionTransformer(log_transform)

# Build a column transformer with numerical preprocessing
preprocessor = ColumnTransformer(
    transformers=[
        ('outliers', outlier_transformer, numerical_cols),
        ('log', log_transformer, numerical_cols),
        ('cat', OneHotEncoder(drop='first'), ['zipcode_processed'])
    ])

```

```python

# Create the full pipeline
pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('regressor', RandomForestRegressor(n_estimators=100))
])

# Assuming X_train, y_train are defined
pipeline.fit(X_train, y_train)
```

The preprocessing for processing outliers and logarithmic transformation had a lower validation metric than the normal model. I will perform Hyperparameter tuning using the regular model  without further preprocessing

# Hyperparameter tuning with Optuna

```python
def objective(trial):
    # Suggesting parameters
    n_estimators = trial.suggest_int('n_estimators', 100, 500)  # more broad range
    max_depth = trial.suggest_int('max_depth', 10, 50)  # allow for potentially deeper trees
    min_samples_split = trial.suggest_int('min_samples_split', 2, 20)
    min_samples_leaf = trial.suggest_int('min_samples_leaf', 1, 10)
    max_features = trial.suggest_categorical('max_features', ['sqrt', 'log2'])

    # Setup RandomForestRegressor with suggested parameters
    regressor = RandomForestRegressor(
        n_estimators=n_estimators,
        max_depth=max_depth,
        min_samples_split=min_samples_split,
        min_samples_leaf=min_samples_leaf,
        max_features=max_features,
        random_state=42
    )
    
    # Creating the pipeline
    forest_pipeline = Pipeline([
        ('preprocessor', preprocessor),
        ('regressor', regressor)
    ])
    
    # Scoring: using negative mean squared error
    score = cross_val_score(forest_pipeline, X_train, y_train, cv=3,
                            scoring='neg_mean_squared_error', n_jobs=-1)
    return score.mean()  # Optuna tries to maximize the objective, hence return negative MSE

# Example usage
study = optuna.create_study(direction='maximize')  # we use 'maximize' because we return negative MSE
study.optimize(objective, n_trials=50)

print("Best trial:", study.best_trial.params)
```

# Refit the Model with Best parameters

```python
best_params = study.best_trial.params
best_forest = RandomForestRegressor(**best_params, random_state=42)
best_pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('regressor', best_forest)
])

best_pipeline.fit(X_train, y_train)

```

# Evaluate the model

```python
y_pred = best_pipeline.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
print("Optimized MSE:", mse)

```
